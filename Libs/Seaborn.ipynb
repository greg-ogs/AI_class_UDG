# Graficando datos con seaborn

## Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Cargar datos

```python
data = pd.read_csv("credit_card.csv")
```

## Creando columna nueva

```python
data['IU'] = data['BILL_AMT'] / data['LIMIT_BAL']
print(data.head(10))  # First 10 rows in python  console
# ages = self.data.AGE.nunique()
# print(ages)
bins = [20, 30, 40, 50, 100]  # Cut in multiple ranges and put in a new column
titles = ['20-30', '30-40', '40-50', '50+']
data['AGE_RANGE'] = pd.cut(data['AGE'], bins, labels=titles)
print('--------------------------------------------------------------------------------------------------')
print(data.head(10))  # First 10 rows in python  console
plt.figure(figsize=(15, 7), facecolor="#E1E4E3", edgecolor="white")  # Figure configuration
plt.title('Unique dispersion data plot using boxplot about age and limit')
sns.set_style('darkgrid')  # Set background color and grid
sns.boxplot(x='AGE_RANGE', y='LIMIT_BAL', data=data, palette='bone', hue='DEFAULT')
plt.show()
```

## Displot para histogramas

```python
plt.figure(figsize=(10, 5), facecolor="white", edgecolor="black")  # Figure configuration
# sns.set_style('whitegrid')
sns.set_style('darkgrid')  # Set background color and grid
sns.displot(data, x='LIMIT_BAL', hue='EDUCATION',
            palette='inferno')  # Create graphics using limit bal column,
# Use hue to add another colum to histogram
# Use palette for color palette
plt.show()
```

## 

```python
plt.figure(figsize=(10, 5), facecolor="white", edgecolor="black")  # Figure configuration
# sns.set_style('whitegrid')
sns.set_style('darkgrid')  # Set background color and grid
sns.displot(data, x='LIMIT_BAL', col='EDUCATION', kind='kde', hue='AGE_RANGE',
            palette='inferno')  # Create graphics using limit bal column,
# kind for graphic style
# 
# line : line plot (default)
# bar : vertical bar plot
# barh : horizontal bar plot
# hist : histogram
# box : boxplot
# kde : Kernel Density Estimation plot
# density : same as kde
# area : area plot
# pie : pie plot
# scatter : scatter plot
# hexbin : hexbin plot
# Use hue to add another colum to histogram,
# Use col to add another plot to histogram and kind for density,
# Use palette for color palette
plt.show()
```

## Countplot para frecuencia

```python
plt.figure(figsize=(10, 5), facecolor="#E1E4E3", edgecolor="white")  # Figure configuration
plt.title('Unique data plot')
sns.set_style('darkgrid')  # Set background color and grid
sns.countplot(x='CHECKING_ACCOUNT', data=data, hue='DEFAULT', palette='inferno_r')
```

## Grafica de dispersion

```python
sns.set_style('darkgrid')  # Set background color and grid
sns.catplot(x='MARRIAGE', y='LIMIT_BAL', data=data, hue='DEFAULT',
            dodge=True)  # Dispersion plot, dodge is used to separate
plt.show()
```

```python
plt.figure(figsize=(50, 25), facecolor="#E1E4E3", edgecolor="white")  # Figure configuration
plt.title('Unique dispersion data plot')
sns.set_style('darkgrid')  # Set background color and grid
sns.swarmplot(x='EDUCATION', y='IU', data=data, palette='inferno_r', hue='DEFAULT')
plt.show()
```

```python
plt.figure(figsize=(50, 25), facecolor="#E1E4E3", edgecolor="white")  # Figure configuration
plt.title('Dispersion plot')
sns.set_style('darkgrid')  # Set background color and grid
sns.scatterplot(x='IU', y='EDUCATION', data=data, palette='inferno', hue='DEFAULT')
plt.show()
```

## Violin

```python
plt.figure(figsize=(50, 25), facecolor="#E1E4E3", edgecolor="white")  # Figure configuration
plt.title('Violin plot')
sns.set_style('darkgrid')  # Set background color and grid
sns.violinplot(x='EDUCATION', y='IU', data=data, palette='inferno', hue='DEFAULT')
plt.show()
```

## lmplot para tendencias en graficas de dispersion

```python
# plt.figure(figsize=(50, 25), facecolor="#E1E4E3", edgecolor="white")  # Figure configuration
# plt.title('Dispersion plot and tendency line using lmplot')
sns.set_style('darkgrid')  # Set background color and grid
sns.lmplot(x='IU', y='LIMIT_BAL', data=data, palette='inferno', hue='DEFAULT', height=9, aspect=1.2)
plt.show()
```

## Graficas bi-variable con grafica mono-variable con jointplot

```python
sns.set_style('darkgrid')  # Set background color and grid
sns.jointplot(x='IU', y='LIMIT_BAL', data=data, palette='inferno', hue='DEFAULT', kind="kde")
sns.jointplot(x='IU', y='LIMIT_BAL', data=data, palette='inferno', kind="hex")
plt.show()
```

## Resumen de el set de datos con pairplot

```python
sns.set_style('darkgrid')  # Set background color and grid
sns.pairplot(data=data, hue="EDUCATION", palette="cividis")
plt.show()
```
